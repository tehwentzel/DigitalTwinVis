```python
data = pd.read_csv('../data/dose_symptoms_merged_paper.csv')
data
```

```python
from ast import literal_eval
for col in data.columns:
    try:
        data[col] = data[col].apply(literal_eval)
    except:
        pass
data['V10'].iloc[0]
```

```python
def get_organ_dvh(df, organ):
    pos = organ_list.index(organ)
    if pos < 0:
        print('bad organ')
        return False
    dvh_cols = [c for c in df.columns if c[0] == 'V' and len(c) < 4]
    dvh = df.apply(lambda x: np.array([xx[pos] for xx in x[dvh_cols]]),axis=1).values
    mean_vals = df['mean_dose'].apply(lambda x: x[pos])
    dvh = np.stack(dvh)
    new_df = pd.DataFrame(dvh,columns=dvh_cols,index=df.index)
    subsite_map = ['BOT','Tonsil']
    new_df['subsite'] = df.subsite.apply(lambda x: x if x in subsite_map else 'other')
    new_df['mean_dose'] = mean_vals
    new_df['organ'] = organ.replace('Rt_','Contra-').replace('Lt_','Ipsi-').replace('_',' ')
    return new_df
    
def get_organs_dvh(df,organs):
    dfs = [get_organ_dvh(df,o).reset_index() for o in organs]
    return pd.concat(dfs,axis=0)
dvh_df = get_organs_dvh(data,['Lt_Parotid_Gland','Rt_Parotid_Gland','Lt_Submandibular_Gland','Rt_Submandibular_Gland'])
dvh_df
```

```python
keys = ['mean_dose','V45','V55']
fig, axes = plt.subplots(len(keys),1,figsize=(10*len(keys),30))
for key,ax in zip(keys,axes):
    sns.boxenplot(
        data=dvh_df,
        x = 'organ',
        y = key,
        hue='subsite',
        palette='colorblind',
        ax=ax,
        dodge=True,
        k_depth='trustworthy',
        hue_order = ['BOT','Tonsil','other']
    )
    title = key.replace('mean_dose','Mean Dose')
    ax.set_title(title,fontsize=32)
    ax.tick_params(labelsize=28)
    ax.set_xlabel('')
    ax.set_ylabel('')
```
