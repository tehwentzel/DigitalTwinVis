```python
from xml.dom import minidom
svgfile = '../resources/ln_diagrams.svg'
doc = minidom.parse(svgfile)
paths = {path.getAttribute('id'): path.getAttribute('d') for path in doc.getElementsByTagName('path') if 'path' not in path.getAttribute('id')}
paths['2A_contra']
```

```python
import json
with open('../resources/ln_diagrams.json','w') as f:
    json.dump(paths,f)
```

```python
with open('../resources/ln_diagrams.json','r') as f:
    test = json.load(f)
test['2A_contra']
```

```python
dltfile = '../resources/dlt_stuff.svg'
dlt = minidom.parse(dltfile)
dlt_dict = {obj.getAttribute('id'): {'d': obj.getAttribute('d'),'style': obj.getAttribute('style')} for obj in dlt.getElementsByTagName('path')}
dlt_dict.keys()
```

```python
with open('../resources/dlt_diagrams.json','w') as f:
    json.dump(dlt_dict,f)
with open('../resources/dlt_diagrams.json','r') as f:
    test = json.load(f)
test
```
